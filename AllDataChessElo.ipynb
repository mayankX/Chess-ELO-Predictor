```
def score_early_mid_end_game(val):
  total_moves = len(val)
  white_early = 0
  black_early = 0
  white_mid = 0
  black_mid = 0
  white_end = 0
  black_end = 0
  early_moves = total_moves/4
  end_moves = total_moves/4
  mid_moves = total_moves/2
  for i in range(1, total_moves):
    move = val[i]
    #Can Fetch other things
    if i <= early_moves:
      if not move['turn']:
        #Means black tunr
        black_early += move['score']
      else:
        white_early += move['score']
    elif i > early_moves and i < early_moves+mid_moves:
      if not move['turn']:
        #Means black tunr
        black_mid += move['score']
      else:
        white_mid += move['score']
    else:
      if not move['turn']:
        #Means black tunr
        black_end += move['score']
      else:
        white_end += move['score']
  return white_early, black_early, white_mid, black_mid, white_end, black_end
  ##The negative score
```

```
def calculate_advantages(val):
  total_moves = len(val)
  white_advantage = 0
  black_advantage = 0
  for i in range(1, total_moves):
    move = val[i]
    #Can Fetch other things
    if not move['turn']:
      #black advantage
      black_advantage += abs(move['score'])
    else:
      white_advantage += abs(move['score'])
  return white_advantage, black_advantage
```

```
game = data.head(551).tail(1)
moves = game["moves"].iloc[0]
game["result"]
```

```
print(score_early_mid_end_game(moves))
calculate_advantages(moves)
```
